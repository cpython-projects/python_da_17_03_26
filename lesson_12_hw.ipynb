### Опис набору даних

Набір даних (**ecommerce_data.csv**, https://raw.githubusercontent.com/cpython-projects/python_da_17_03_26/refs/heads/main/ecommerce.csv) містить **40 рядків** і **19 стовпців**, що представляють транзакції у сфері електронної комерції. Нижче наведено опис кожного стовпця:

| Назва стовпця      | Опис                                                                       |
| ------------------ | -------------------------------------------------------------------------- |
| `order_id`         | Унікальний ідентифікатор замовлення.                                       |
| `customer_id`      | Унікальний ідентифікатор клієнта.                                          |
| `order_date`       | Дата оформлення замовлення (у форматі РРРР-ММ-ДД).                         |
| `product_id`       | Унікальний ідентифікатор придбаного товару.                                |
| `product_name`     | Назва товару.                                                              |
| `category`         | Категорія товару (наприклад, Електроніка, Одяг тощо).                      |
| `price`            | Ціна за одиницю товару.                                                    |
| `quantity`         | Кількість придбаних одиниць.                                               |
| `weight`           | Вага товару (у кг, зберігається як рядок).                                 |
| `discount`         | Застосована знижка (у вигляді десяткового дробу, наприклад, 0.10 для 10%). |
| `shipping_cost`    | Вартість доставки товару (наявні пропущені значення).                      |
| `payment_method`   | Метод оплати (наприклад, Кредитна картка, PayPal, Дебет).                  |
| `delivery_status`  | Статус доставки (Доставлено, Відправлено, В обробці тощо).                 |
| `customer_city`    | Місто клієнта.                                                             |
| `customer_state`   | Штат клієнта.                                                              |
| `customer_country` | Країна клієнта.                                                            |
| `return_requested` | Чи було запитано повернення (1 = Так, 0 = Ні).                             |
| `review_score`     | Оцінка клієнта (за 5-бальною шкалою, є пропущені значення).                |
| `days_to_deliver`  | Кількість днів на доставку (є пропущені значення).                         |

**Завдання 2.** Обробка некоректних форматів

* Числові дані подані у вигляді рядків
* Дати подані у вигляді рядків
* Категоріальні дані подані у вигляді рядків/чисел


```python
df.info()
```

```python
df.columns.to_list()
```

```python
df.order_date.unique()
```

```python
df['order_date'] = pd.to_datetime(df.order_date, format='%Y-%m-%d', errors='coerce')
```

```python
df['category'] = df.category.astype('category')
```

```python
df['payment_method'] = df.payment_method.str.strip().str.lower()
```

```python
df.payment_method.replace(
    {
        'credit card': 'credit',
        'debit card': 'debit'
    },
    inplace=True
)
```

```python
df['payment_method'] = df.payment_method.astype('category')
```

```python
df['delivery_status'] = df.delivery_status.str.lower()
```

```python
for column_name in []:
  df[column_name] = df[column_name].str.lower()
```

```python
df['delivery_status'] = df.delivery_status.astype('category')
```

```python
def weight_converter(item):
  item = item.lower()
  if 'kg' in item:
    return float(item.replace('kg', ''))
  if 'lbs' in item:
    return float(item.replace('lbs', '')) * 0.45

  return None


df['weight_kg'] = df.weight.apply(weight_converter)
```

```python
df.info()
```

**Завдання 3.** Робота з пропущеними значеннями

```python
# discount
# shipping_cost
# review_score
# days_to_deliver
df['discount'] = df.discount.fillna(0)
```

```python
df[~df['shipping_cost'].notna()]
```

```python
df['review_score '] = df.review_score.fillna(df.review_score.median())
```

**Завдання 4.** Визначення найприбутковіших товарів

**Необхідні ознаки:** `product_name`, `price`, `quantity`, `discount`

```python
df['total'] = df.price * df.quantity * (1 - df.discount)
```

```python
df.sort_values(by='total', ascending=False, inplace=True)
```
