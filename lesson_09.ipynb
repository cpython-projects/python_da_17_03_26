```python
customer_purchases = [
    {"Customer_ID": 1001, "Product": "Laptop", "Category": "Electronics", "Purchase_Date": "2024-02-01"},
    {"Customer_ID": 1002, "Product": "Smartphone", "Category": "Electronics", "Purchase_Date": "2024-02-02"},
    {"Customer_ID": 1003, "Product": "Sofa", "Category": "Furniture", "Purchase_Date": "2024-02-03"},
    {"Customer_ID": 1001, "Product": "Tablet", "Category": "Electronics", "Purchase_Date": "2024-02-10"},
    {"Customer_ID": 1002, "Product": "Chair", "Category": "Furniture", "Purchase_Date": "2024-02-15"},
    {"Customer_ID": 1003, "Product": "Laptop", "Category": "Electronics", "Purchase_Date": "2024-02-18"},
    {"Customer_ID": 1004, "Product": "Desk", "Category": "Furniture", "Purchase_Date": "2024-02-20"},
]
```

Визначте множину всіх унікальних товарів, які були куплені.

Виведіть список унікальних Customer_ID, які купили хоча б один товар із категорії "Electronics".

```python
res = []


for order in customer_purchases:
  if order["Product"] not in res:
    res.append(order["Product"])

print(res)
```

```python
res = {order["Product"] for order in customer_purchases}
print(res)
```

```python
res = {order["Customer_ID"] for order in customer_purchases if order['Category'] == 'Electronics'}
print(res)
```

```python
res = set()

for order in customer_purchases:
  if order['Category'] == 'Electronics':
    res.add(order["Customer_ID"])

print(res)
```

```python
f = open('text.txt', 'w', encoding='utf-8')  # r - read, w - write, a - append, r+ - read + write

f.write('Hello\n')
f.write('world')

f.close()
```

```python
with open('text.txt', 'w', encoding='utf-8') as f:
  f.write('Hello\n')
  f.write('world')
```

```python
with open('text.txt', 'r', encoding='utf-8') as f:
  data = f.read()

  print(data)
```

```python
with open('text.txt', 'r', encoding='utf-8') as f:
  data = f.readlines()

  print(data)
```

```python
with open('text.txt', 'r', encoding='utf-8') as f:
  for line in f:
    print(line.strip())
```
