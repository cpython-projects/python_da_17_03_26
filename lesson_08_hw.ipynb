**Task 1.** Напишіть скрипт, який приймає рядок тексту і повертає словник, де ключами є слова, а значеннями – кількість входжень кожного слова в тексті.

````python
import string
from collections import Counter

text = 'World is big, but my Own world is small'
for item in string.punctuation:
  text = text.replace(item, ' ')
text = text.lower().split()
result = Counter(text)
result
````

**Task 4.** Дано список клієнтів, кожен клієнт представлений словником з полями: `"id"`, `"ім'я"`, `"вік"`, `"місто"`.

```python
[
    {"id": 1, "ім'я": "Анна", "вік": 25, "місто": "Берлін"},
    {"id": 2, "ім'я": "Іван", "вік": 30, "місто": "Берлін"},
    {"id": 3, "ім'я": "Ольга", "вік": 22, "місто": "Відень"}
]
```

Напишіть скрипт, який формує словник, у якому ключі — міста, а значення — клієнти (список) з цих міст.

````python
from collections import defaultdict

data = [
    {"id": 1, "ім'я": "Анна", "вік": 25, "місто": "Берлін"},
    {"id": 2, "ім'я": "Іван", "вік": 30, "місто": "Берлін"},
    {"id": 3, "ім'я": "Ольга", "вік": 22, "місто": "Відень"}
]

result = defaultdict(list)

for rows in data:
  result[rows['місто']].append(rows["ім'я"])

result
````

**Task 5.** Дано список працівників, де кожен працівник представлений словником з полями `"ім'я"`, `"відділ"`.

```python
[
    {"ім'я": "Анна", "відділ": "Маркетинг"},
    {"ім'я": "Іван", "відділ": "Продажі"},
    {"ім'я": "Ольга", "відділ": "Маркетинг"}
]
```

Створіть скрипт, який групує працівників за відділами.

````python
from collections import defaultdict

data = [
    {"ім'я": "Анна", "відділ": "Маркетинг"},
    {"ім'я": "Іван", "відділ": "Продажі"},
    {"ім'я": "Ольга", "відділ": "Маркетинг"}
]

result = defaultdict(list)

for rows in data:
  result[rows['відділ']].append(rows["ім'я"])

result
````

**Task 6.** Дано словник з даними щодо продажів, де ключі — дні місяця, а значення — сума продажу.

```python
{
    1: 500, 2: 700, 3: 300, 4: 800, 5: 200
}
```

Знайдіть дні з найбільшими відхиленнями від середнього значення.

````python
import statistics

data = {
    1: 500, 2: 700, 3: 300, 4: 800, 5: 200
}

avg = statistics.mean(data.values())
avg
````

````python
diff = [abs(value - avg) for value in data.values()]
diff
````

````python
max_value = max(diff)
max_value
````

````python
import math

for i, item in enumerate(diff):
  if math.isclose(item, max_value):
      print(i + 1)
````
