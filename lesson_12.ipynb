```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cpython-projects/python_da_17_03_26/refs/heads/main/lesson_12_data.csv')
df
```

# STEP 1

```python
df.columns.to_list()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
df.rename(columns={
    'ID': 'id',
    'Name': 'employee',
    'Age': 'age',
    'Salary ($)': 'employee_salary',
    'Hire Date': 'hire_date',
    'Country': 'country',
    'Weight': 'employee_weight'
}, inplace=True)
df
```

```python
df.info()
```

# STEP 2

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

# STEP 3

```python
df['employee'] = df.employee.str.title()
df
```

```python
df.gender.unique()
```

```python
df['gender'] = df.gender.astype('category')
```

```python
df.info()
```

```python
df.country.unique()
```

```python
df['country'] = df.country.str.title()
```

```python
df['country'] = df.country.replace({
    'Usa': 'U.S.A.',
    'Uk': 'United Kingdom',
})
```

```python
df
```

```python
df.age.unique()
```

```python
df['age'] = df.age.replace({
    'Thirty': '30',
    'Twenty': '20'
})
```

```python
df['age'] = df.age.astype('int')
```

```python
df['age'] = pd.to_numeric(df.age, errors='coerce')
df
```

```python
df['age'] = df.age.fillna(df.age.mean())
df['age'] = df.age.astype('int')
df
```

```python
df['new_date'] = pd.to_datetime(df.hire_date, errors='coerce')
df
```

```python
from dateutil import parser
```

```python
df['hire_date'] = df.hire_date.apply(parser.parse)
df
```

```python
df.info()
```

```python
def convert_weight(item):
  item = item.lower().strip()
  if 'kg' in item:
    return float(item.replace('kg', ''))

  if 'lbs' in item:
    return float(item.replace('lbs', '')) * 0.45345678

  return float(item)

df['weight_kg'] = df.employee_weight.apply(convert_weight)
df
```
