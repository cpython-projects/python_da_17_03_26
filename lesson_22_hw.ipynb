# Task. Спроєктувати базу даних для мобільного застосунку з продажу товарів

# 📁 Сутності:

* Встановлення застосунку (`app_sessions`)
* Перегляди товарів (`product_views_log`)
* Користувачі та пристрої (`devices_users_map`)
* Замовлення (`orders_log`)

---

## 📁 Поля:

### 1. Таблиця `app_sessions`
* `session_id`: унікальний ідентифікатор сесії
* `device_code`: код пристрою
* `first_seen`: дата першого запуску
* `os_type`: платформа (Android або iOS)
* `acquisition_channel`: канал залучення (наприклад, Facebook, Organic тощо)
* `cpi_uah`: вартість установки в гривнях

### 2. Таблиця `product_views_log` — перегляди товарів
* `device_code`: код пристрою
* `view_date`: дата перегляду
* `platform`: платформа
* `view_count`: кількість переглядів за день

### 3. Таблиця `devices_users_map`
* `device_code`: код пристрою
* `user_uuid`: унікальний ID користувача (присвоюється після авторизації)

> Зверніть увагу: не всі користувачі авторизуються

### 4. Таблиця `orders_log`
* `user_uuid`: ID користувача
* `order_time`: дата замовлення
* `total_uah`: сума покупки в гривнях

---

### 📁 Зв’язки:

* Один пристрій (`device_code`) може встановити застосунок лише один раз
* Один пристрій може переглядати товари багаторазово
* Один користувач може бути прив'язаний до кількох пристроїв, але в системі зафіксований через `user_uuid`
* Один користувач може зробити кілька замовлень
* Перегляди товарів відбуваються до авторизації, тому `device_code` пов'язує встановлення, перегляди та авторизовані покупки
