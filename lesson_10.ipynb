```python
x = [12, 23, 34, 43, -12, -34, 12, 56, 67]
```

```python
def filter_even(seq):
  result = []

  for item in seq:
    if item % 2 == 0:
      result.append(item)

  return result

filter_even(x)
```

```python
def filter_neg(seq):
  result = []

  for item in seq:
    if item < 0:
      result.append(item)

  return result

filter_neg(x)
```

```python
def filter_odd(seq):
  result = []

  for item in seq:
    if item % 2:
      result.append(item)

  return result

filter_odd(x)
```

```python
def filter_items(seq, func):
  res = []

  for item in seq:
    if func(item):
      res.append(item)
  return res
```

```python
x = [12, 23, 34, 43, -12, -34, 12, 56, 67]
```

```python
def is_odd(item):
  return item % 2 == 1

def is_even(item):
  return item % 2 == 0

def is_positive(item):
  return item > 0
```

```python
filter_items(x, is_odd)
```

```python
filter_items(x, is_even)
```

```python
filter_items(x, is_positive)
```

```python
prices = ['12$', '23$', '78$', '34', '78$']
products = ['banana', 'apple', 'kiwi', 'grape', 'pineapple']

def clean_price(item):
  return float(item.replace('$', ''))

clean_ = [clean_price(item) for item in prices]
print(clean_price)
```

```python
clean_ = map(clean_price, prices)
print(list(clean_))
```

```python
clean_ = zip(
    map(
        lambda item: float(item.replace('$', '')),
        prices
      ),
    products
  )
print(list(clean_))
```
