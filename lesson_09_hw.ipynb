Task 4. Знайдіть клієнтів, які купували товари лише з однієї категорії.

````python
from collections import defaultdict

users_grouped = defaultdict(set)

for order in customer_purchases:
  users_grouped[order["Customer_ID"]].add(order['Category'])
users_grouped
````

````python
for key, value in users_grouped.items():
  if len(value) == 1:
    print(key)
````

````python
for item in enumerate(users_grouped):
  print(item)
````

Task 7. Для кожного клієнта визначте товари, які він купив, але ніхто інший із покупців не купував.

````python
users = {order['Customer_ID'] for order in customer_purchases }
users
````

````python
for user in users:
  cart_user = set()
  cart_others = set()


  for order in customer_purchases:
    if order['Customer_ID'] == user:
      cart_user.add(order['Product'])
    else:
      cart_others.add(order['Product'])
  user_without_others = cart_user - cart_others
  print(user, user_without_others)
````
