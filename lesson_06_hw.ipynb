**Task 4.** Користувач вводить з клавіатури текст. Знайдіть у ньому найдовше слово і виведіть його на екран.

```python
import string
text = input('text>>').lower()

for item in string.punctuation:
  text = text.replace(item, ' ')

words = text.split()
words
```

**Task 6.** Дано рядок із текстом, що містить хештеги (наприклад, "#data #Python"). Видобудьте їх у список.

```python
text = '#data #Python'
```

```python
res = text.split('#')
while '' in res:
  res.remove('')

for index, word in enumerate(res):
  res[index] = word.strip()

res
```

**Task 7.** Дано рядок з текстом. Залиште тільки слова довжиною понад 4 символи.

```python
x = 'Hello world Python is awesome'.lower()

words = x.split()

result = [word for word in words if len(word) > 4]

print(result)
print(' '.join(result))
```

**Task 8.** У рядку замініть усі цифри на символ \*.

```python
import string

text = 'My phone number is 1234567890'

for i in string.digits: # for i in '0123456789'
  text = text.replace(i, '*')
print(text)
```

**Task 10.** У рядку видаліть усі повторювані слова, зберігши їхній порядок.

```python
x = 'Hello WORLD Python is awesome Python is'
x = x.lower()

res = []


for word in x.split():
  if x.count(word) == 1:
    res.append(word)

print(' '.join(res))
```
