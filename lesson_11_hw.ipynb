**Task 7.**

```python
df_filtered.FamilySize.unique()
```

```python
for item in sorted(df_filtered.FamilySize.unique()):
  tmp = df_filtered[df_filtered['FamilySize'] == item]
  print(f'Size: {item}, {tmp.Survived.mean() * 100:.2f}%')
```

**Task 8.**
Чи були відмінності у виживанні серед пасажирів з різними титулами (Mr, Mrs, Dr тощо)?

```python
def get_person_title(name):
  name = name.split()
  for part in name:
    if part[-1] == '.':
      return part
  return ''

df_filtered['Titule'] = df_filtered.Name.apply(get_person_title)
df_filtered
```
