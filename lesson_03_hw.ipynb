**Task 6.** Користувач вводить три числа (дохід, витрати, потрібна сума кредиту). Визначте, чи зможе людина сплатити кредит за N місяців.

```python
import math
salary = float(input('salary>>'))
expenses = float(input('expenses>>'))
credit = float(input('credit>>'))
months = int(input('months>>'))


free = salary - expenses
res = 'Yes' if math.ceil(credit / free) <= months else 'No'
print(res)
```

**Task 7.** Користувач вводить два числа (масу та зріст). Обчисліть індекс маси тіла (ІМТ) та визначте категорію (норма, надмірна вага тощо).

```python
weight, height = float(input('weight>>')), float(input('height (m)>>'))

x = weight / (height ** 2)

if x < 18.5:
  res = 'Недостатня'
elif x < 25:
  res = 'Норма'
elif  x <= 30:
  res = 'Надмірна'
else:
  res = 'Ожиріння'

print(res)
```
